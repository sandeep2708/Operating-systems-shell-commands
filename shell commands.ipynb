```python
import subprocess
import platform

def execute_command(command):
    try:
        result = subprocess.run(command, shell=True, check=True, capture_output=True, text=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        raise RuntimeError(f"Error executing command: {e}\nCommand: {e.cmd}\nOutput: {e.output}")

def print_command_result(output):
    print("Command executed successfully:")
    print(output)

def menu():
    print("\nExample Menu:")
    print("1. List directory contents")
    print("2. Print working directory")
    print("3. Create a new directory")
    print("4. Display a message")
    print("5. Concatenate and display content of a file")
    print("6. Exit")

def main():
    while True:
        menu()
        option = input("Enter any option number from (1-6): ")

        if option == "1":
            # Use 'ls' for both Unix-like systems and Windows
            try:
                result = execute_command("ls")
                print_command_result(result)
            except RuntimeError as e:
                print(e)
        elif option == "2":
            try:
                result = execute_command("pwd" if "posix" in platform.system().lower() else "cd")
                print_command_result(result)
            except RuntimeError as e:
                print(e)
        elif option == "3":
            dir_name = input(" Please enter the new directory name to be created: ")
            try:
                execute_command(f"mkdir {dir_name}")
                print("Directory created successfully.")
            except RuntimeError as e:
                print(e)
        elif option == "4":
            message = input("Please enter the message to be displayed: ")
            print(message)
        elif option == "5":
            file_name = input("Enter the file name: ")
            try:
                result = execute_command(f"cat {file_name}" if "posix" in platform.system().lower() else f"type {file_name}")
                print_command_result(result)
            except RuntimeError as e:
                print(e)
        elif option == "6":
            print("Exiting the program.")
            break
        else:
            print("The option is invalid. Please enter a number from 1 and 6.")

if __name__ == "__main__":
    main()
```
